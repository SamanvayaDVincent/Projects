3) Logistic Regression:

a. Build a simple logistic regression model:

i. Divide the dataset in 70:30 ratio

ii. Build the model on train set and predict the values on test set

iii. Build the confusion matrix and get the accuracy score

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score

x = heart.drop('target',axis=1)
y = heart['target']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 40)

hrt_dis = LogisticRegression()
hrt_dis.fit(x_train,y_train)
print('Trained Model: ',hrt_dis)

y_pred = hrt_dis.predict(x_test)
print('Predicted values: ', y_pred)

con_mat = confusion_matrix(y_test,y_pred)
print('Confusion Matrix:',con_mat)

accuracy = accuracy_score(y_test,y_pred)
print('Accuracy Score: ', accuracy)
```

4) Decision Tree:

a. Build a decision tree model:

i. Divide the dataset in 70:30 ratio

ii. Build the model on train set and predict the values on test set

iii. Build the confusion matrix and calculate the accuracy

iv. Visualize the decision tree using the Graphviz package

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
import numpy as np


x.columns = [str(i) for i in x.columns]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state = 40)

hrt_dis1 = DecisionTreeClassifier()
y_pred1 = hrt_dis1.fit(x_train,y_train).predict(x_test)
print('Predicted values: ', y_pred1)

con_mat1 = confusion_matrix(y_test,y_pred1)
print('Confusion Matrix:',con_mat1)

accuracy1 = accuracy_score(y_test,y_pred1)
print('Accuracy Score: ', accuracy1)

heart_graph = export_graphviz(hrt_dis1)
graph = graphviz.Source(heart_graph)
graph.render('Decision Tree')
```

5) Random Forest:

a. Build a Random Forest model:

i. Divide the dataset in 70:30 ratio

ii. Build the model on train set and predict the values on test set

iii. Build the confusion matrix and calculate the accuracy

iv. Visualize the model using the Graphviz package

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import graphviz

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=40)

rf_heart = RandomForestClassifier(n_estimators = 100, random_state=40)
y_pred = rf_heart.fit(x_train,y_train).predict(x_test)
print("predicted values: ", y_pred)

conf_mat = confusion_matrix(y_test,y_pred)
print('Confusion Matrix: ',conf_mat)

accuracy = accuracy_score(y_test,y_pred)
print("Acuuracy Score is: ",accuracy)

rf_graph = graphviz.Source(export_graphviz(rf_heart.estimators_[0]))
rf_graph.format='png'
rf_graph.render('Random Forest')
rf_graph.view()

```

6) Select the best model

a. Print the confusion matrix of all classifiers

b. Print the classification report of all classifiers

c. Calculate Recall Precision and F1 score of all the models

d. Visualize confusion matrix using heatmaps

e. Select the best model based on the best accuracies

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=40)

dt_model = DecisionTreeClassifier(random_state=42, max_depth=5)
dt_pred = dt_model.fit(x_train, y_train).predict(x_test)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_pred = rf_model.fit(x_train, y_train).predict(x_test)

nb_model = GaussianNB()
nb_pred = nb_model.fit(x_train, y_train).predict(x_test)

#confusion matrix
print('Confusion Matrix (Decision Tree):\n', confusion_matrix(y_test, dt_pred))
print('Confusion Matrix (Random Forest):\n', confusion_matrix(y_test, rf_pred))
print('Confusion Matrix (Naive Bayes):\n', confusion_matrix(y_test, nb_pred))

#classification report
print('Classification Report (Decision Tree):\n', classification_report(y_test, dt_pred))
print('Classification Report (Random Forest):\n', classification_report(y_test, rf_pred))
print('Classification Report (Naive Bayes):\n', classification_report(y_test, nb_pred))

#recall score
print('Recall (Decision Tree):', recall_score(y_test, dt_pred))
print('Recall (Random Forest):', recall_score(y_test, rf_pred))
print('Recall (Naive Bayes):', recall_score(y_test, nb_pred))

#precision score
print('Precision (Decision Tree):', precision_score(y_test, dt_pred))
print('Precision (Random Forest):', precision_score(y_test, rf_pred))
print('Precision (Naive Bayes):', precision_score(y_test, nb_pred))

#F1 score
print('F1 Score (Decision Tree):', f1_score(y_test, dt_pred))
print('F1 Score (Random Forest):', f1_score(y_test, rf_pred))
print('F1 Score (Naive Bayes):', f1_score(y_test, nb_pred))

#heatmap
fig, axs = plt.subplots(ncols=3, figsize=(15, 5))
sns.heatmap(confusion_matrix(y_test, dt_pred), ax=axs[0], annot=True, cmap='Blues')
sns.heatmap(confusion_matrix(y_test, rf_pred), ax=axs[1], annot=True, cmap='Blues')
sns.heatmap(confusion_matrix(y_test, nb_pred), ax=axs[2], annot=True, cmap='Blues')
axs[0].set_title('Decision Tree')
axs[1].set_title('Random Forest')
axs[2].set_title('Naive Bayes')
plt.show()

#accuracy
dt_accuracy = accuracy_score(y_test, dt_pred)
rf_accuracy = accuracy_score(y_test, rf_pred)
nb_accuracy = accuracy_score(y_test, nb_pred)

#select best method
if dt_accuracy > rf_accuracy and dt_accuracy > nb_accuracy:
    print('Decision Tree has the highest accuracy:', dt_accuracy)
elif rf_accuracy > dt_accuracy and rf_accuracy > nb_accuracy:
    print('Random Forest has the highest accuracy:', rf_accuracy)
else:
    print('Naive Bayes has the highest accuracy:', nb_accuracy)
```
