3) From the dataset, calculate the number of genuine transactions, number of fraud transactions and the percentage of fraud transactions.


```python
gen_trans = credit['Class'].value_counts()[0]
fraud_trans = credit['Class'].value_counts()[1]
perc_fraudtr = (fraud_trans / (fraud_trans + gen_trans))*100

print("Genuine Transactions: ", gen_trans)
print("Fraud Transcations: ",fraud_trans)
print("Percentage of Fraud Transcations: %",format(perc_fraudtr))
```

4) Using the visualization module, visualize the genuine and fraudulent transactions using a bar graph.


```python
import pandas as pd
import matplotlib.pyplot as plt

gen_trans = credit['Class'].value_counts()[0]

fraud_trans = credit['Class'].value_counts()[1]

plt.bar(['Genuine', 'Fraud'], [gen_trans, fraud_trans])

plt.title('Number of Genuine and Fraudulent Transactions')
plt.xlabel('Transaction Type')
plt.ylabel('Number of Transactions')

plt.show()
```

5) Using the Standard Scaler module, normalize the amount column and
store the new values in the NormalizedAmount column.


```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
credit['NormalizedAmount']=scale.fit_transform(credit['Amount'].values.reshape(-1,1))
credit.drop(['Amount','Time'], inplace=True,axis=1)
credit
```

6) Split the dataset in train and test set and have a 70:30 split ratio for the
model.


```python
from sklearn.model_selection import train_test_split

x = credit.drop(['Class'],axis=1)
y = credit['Class']
x_train, y_train,x_test,y_test= train_test_split(x,y, test_size=0.3, random_state=42)

print(x_train.shape,y_train.shape)

x_train = x.iloc[:8164, :]
y_train = y.iloc[:8164]

print(x_train.shape,y_train.shape)
```

7) Now use a decision tree and random forest model for training on top of the
train set.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
print('Decision Tree: ',dt_model)

rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(x_train,y_train)
print('Random Forest: ',rf_model)
```

8) Compare the predictions of both models using predict().


```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, recall_score, f1_score
import numpy as np
print('x_test shape: ',x_test.shape)
x_test = x.iloc[:31, :]

dt_preds = dt_model.predict(x_test)
rf_preds = rf_model.predict(x_test)

print('Decision Tree Prediction: ',dt_preds)
print('Random Forest Prediction: ',rf_preds)
```

9) Compare the accuracy of both models using score().


```python
print('y_test shape:', y_test)
y_test = y.iloc[:31]
dt_score = accuracy_score(y_test, dt_preds)
rf_score = accuracy_score(y_test, rf_preds)
dt_score = precision_score(y_test, dt_preds)
rf_score = precision_score(y_test, rf_preds)
dt_score = recall_score(y_test, dt_preds)
rf_score = recall_score(y_test, rf_preds)
dt_score = f1_score(y_test, dt_preds)
rf_score = f1_score(y_test, rf_preds)
print('Decision tree model accuracy score:', dt_score)
print('Random forest model accuracy score:', rf_score)
print('Decision tree precision score:', dt_score)
print('Random forest precision score:', rf_score)
print('Decision tree recall score:', dt_score)
print('Random forest recall score:', rf_score)
print('Decision tree f1 score:', dt_score)
print('Random forest f1 score:', rf_score)
```

10) Check the performance matrix of both models and compare which
model is having the highest performance.

```python
dt_cm = confusion_matrix(y_test, dt_preds)
rf_cm = confusion_matrix(y_test, rf_preds)
print('Decision tree model confusion matrix:')
print(dt_cm)
print('Random forest model confusion matrix:')
print(rf_cm)
```
